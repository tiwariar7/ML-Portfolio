```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.axes as ax
```

<span style="font-size:16px; color:blue;">Load DataSet do Pre-processing</span> 

```python
data = pd.read_csv("data_for_lr.csv")
data.head(10)
```

```python
data.isnull().sum()
```

<span style="font-size:16px; color:blue;">Data Processing</span> 

```python
data = data.dropna()
print(data.shape)
```

<span style="font-size:16px; color:blue;">Splitting the Data</span>

```python
train_input = np.array(data.x[0:500]).reshape(500,1)
train_output = np.array(data.x[0:500]).reshape(500,1)
```

<span style="font-size:16px; color:blue;">Validating Input Output</span>

```python
test_input = np.array(data.x[500:700]).reshape(199,1)
test_output = np.array(data.x[500:700]).reshape(199,1)
```

<span style="font-size:22px; color:green;">Linear Regression</span>

<span style="font-size:20px; color:red;">Step 1 : Forward Propagation</span> <br>
<span style="font-size:16px; color:blue;">y = mx + c</span>

```python
def forward_propagation(train_input, parameters):
    """Compute the predicted values (y_hat) using the current parameters."""
    m = parameters["m"]
    c = parameters["c"]
    prediction = np.multiply(m, train_input) + c
    return prediction
```

<span style="font-size:20px; color:red;"> Step 2: Calculate Error </span><br>
<span style="font-size:16px; color:blue;">cost function = 1/2 * ∑(y' - y)^2</span>

```python
def calculate_error(predictions,train_output):
    """Compute the Mean Squared Error (MSE) between actual and predicted values."""
    cost = np.mean((train_output - predictions) ** 2) * 1/2
    return cost
```

<span style="font-size:20px; color:red;">Step 3: Gradient Descent Algorithm {Backward Propagation}</span><br>
<span style="font-size:16px; color:blue;">m1 = m - α * dm </span><br>
<span style="font-size:16px; color:blue;">c1 = c - α * dc </span><br>
<span style="font-size:16px; color:blue;">df = ∂cost/∂m = (y1 - y)/n </span><br>
<span style="font-size:16px; color:blue;">dm = x * ∂cost/∂m = (y1 - y)/n = df * x </span><br>
<span style="font-size:16px; color:blue;">dc = ∂cost/∂m = (y1 - y)/n = df </span>

```python
def gradient_descent(train_input, train_output, predictions):
    """Compute gradients for m and c."""
    derivatives = dict()
    df = predictions - train_output
    dm = np.mean(np.multiply(df, train_input))
    dc = np.mean(df)
    derivatives["dm"] = dm
    derivatives["dc"] = dc
    return derivatives
```

<span style="font-size:16px; color:blue;">Update Parameters</span>

```python
def update_parameters(parameters, derivatives, learning_rate):
    """Update the parameters w and b using the gradients and learning rate."""
    parameters["m"] = parameters["m"] - learning_rate * derivatives["dm"]
    parameters["c"] = parameters["c"] - learning_rate * derivatives["dc"]
    return parameters
```

<span style="font-size:22px; color:red;">Model Training</span>

```python
def train(train_input, train_output, learning_rate, iters):
    """Train the linear regression model."""
    # Initialize random parameters
    parameters = dict()
    parameters["m"] = np.random.uniform(0, 1)
    parameters["c"] = np.random.uniform(0, 1)

    plt.figure(figsize=(4, 2))

    #loss
    loss = list()
    
    # Iterate to optimize parameters
    for i in range(iters):
        # Forward propagation
        predictions = forward_propagation(train_input, parameters)
        # Calculate cost
        cost = calculate_error(predictions, train_output)

        loss.append(cost)
        print(f"Iteration = {i+1}, Loss = {cost}")

        #plot
        fig, ax = plt.subplots()

        ax.plot(train_input, train_output, "+", label = "Original")
        ax.plot(train_input, predictions, "*", label = "Training")

        legend = ax.legend()

        plt.plot(train_input, train_output, "+", label = "Original")
        plt.plot(train_input, predictions, "*", label = "Training")
        plt.show()
        
        # Backward propagation
        derivatives = gradient_descent(train_input, train_output, predictions)
        # Update parameters
        parameters = update_parameters(parameters, derivatives, learning_rate)
        
        # Print cost every 100 iterations
        #if i % 100 == 0:
         #   print(f"Iteration {i}, Cost: {cost}")
    
    return parameters, loss
```

```python
parameters, loss = train(train_input,train_output,0.1,20)
```
